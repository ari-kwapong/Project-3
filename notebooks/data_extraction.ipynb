```python
# IMPORT COMMAND:
# 'mongoimport --type json -d crime -c toronto_crime --drop --jsonArray Neighbourhood_Crime_Rates_Open_Data_chopped.geojson'

# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd

# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python

# confirm that our new database was created
print(mongo.list_database_names())
```

```python
db = mongo['crime']
# review the collections in our new database
print(db.list_collection_names())
```

```python
# review a document in the establishments collection
pprint(db.toronto_crime.find_one())
```

```python
results = list(db.toronto_crime.find())
# Convert the result to a Pandas DataFrame
df = pd.json_normalize(results)
# Display the number of rows in the DataFrame
print(df.shape[0])
# Display the first 10 rows of the DataFrame
df.head(10)
```

```python
# One way to iterate through the dataframe
for i in range(len(df)):
    print(df.loc[i, "properties.AREA_NAME"])
```
